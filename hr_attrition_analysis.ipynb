```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Downloads/Sample data for project/HR_Employee_Attrition_Analysis.csv')
df.head()
```

```python
# Step 1: Check Dataset Shape:
df.shape
```

```python
# Step 2: Dataset Information
df.info
```

```python
# Step 3: Check Missing Values:
df.isnull().sum()
```

```python
# Step 4: Attrition Distribution:

import seaborn as sns

sns.countplot(x='Attrition', data=df)
plt.title('Employee Attrition Distribution')
plt.show()
```

```python
# Step 5: Department-wise Attrition

pd.crosstab(df["Department"], df["Attrition"]).plot(
    kind = "bar",
    stacked="true",
    figsize= (8, 5)
)
plt.title("Department-wise Attrition")
plt.ylabel("Number of Employee")
plt.show()
```

```python
# Step 6: Overtime vs Attrition:
sns.countplot(x='OverTime', hue='Attrition', data=df)

plt.title('Overtime vs Attrition')
plt.show()
```

```python
# Analysis: Age vs Attrition

plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='Age',
    hue='Attrition',
    multiple='stack'
)

plt.title('Age Distribution by Attrition')
plt.show()
```

```python
# Monthly Income vs Attrition:

plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df
)

plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
# Job Role vs Attrition

plt.figure(figsize=(12,6))

sns.countplot(
    x='JobRole',
    hue='Attrition',
    data=df
)

plt.xticks(rotation=45)
plt.title('Job Role vs Attrition')
plt.show()
```

```python
# Correlation Heatmap:

numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(14,10))
sns.heatmap(numeric_df.corr())
plt.title('Correlation Heatmap')
plt.show()
```
