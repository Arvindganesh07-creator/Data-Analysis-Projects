```python
ipl["runs"] = pd.to_numeric(ipl["runs"], errors = "coerce")
ipl["runs"].dtype
```

```python
# Find Top Run Scorers
top_runs = (
    ipl.groupby("batsman")["runs"]
    .sum()
    .sort_values(ascending=False)
    .head()
)
print(top_runs)
```

### Insight:
Virat Kohli is the highest run scorer in the dataset with 5691 runs,
followed by David Warner and Suresh Raina.

```python
# Create a Bar Chart
import matplotlib.pyplot as plt

top_runs.plot(kind ="bar")

plt.title("top 5 IPL runs score")
plt.xlabel("batsman")
plt.ylabel("totl runs")

plt.show()
```

```python
# Top Wicket Takers:
ipl["wicketPlayerOut"].notna().sum()
```

```python
# Calculate Top Wicket Takers:
top_wicket =(
    ipl.groupby("bowler")["wicketPlayerOut"]
    .count()
    .sort_values(ascending = False)
    .head(10)
)
print(top_wicket)
```

## Analysis 2: Top Wicket Takers

### Objective
Identify the bowlers who have taken the most wickets in IPL history.

### Insight
Harbhajan Singh leads the wicket charts, followed by R Ashwin and PP Chawla.
These bowlers have consistently performed across multiple IPL seasons.

```python
import matplotlib.pyplot as plt

top_wicket.plot(kind = "bar", figsize = (8,5))

plt.title("Top 10 IPL wicket taker")
plt.xlabel("bowler")
plt.ylabel("wicket")

plt.show()
```

```python
# Winning Teams:

winning_team = (
    ipl ["winner"]
    .value_counts()
    .head(10)
)
print(winning_team)
```

```python
import matplotlib.pyplot as plt

winning_team.plot (kind = "bar", figsize = (8,5))

plt.title("IPL top winning team")
plt.xlabel("team")
plt.ylabel("win")
plt.xticks(rotation = 45)
plt.tight_layout()

plt.show()
```

## Analysis 3: Winning Teams

```python
# Check Venue Data:

ipl["venue"].head()
```

```python
# Count Matches by Venue

venue_count =(
    ipl ["venue"]
    .value_counts()
    .head(10)
)
print(venue_count)
```

```python
# Create Visualization

import matplotlib.pyplot as plt

venue_count.plot(kind = "bar", figsize= (10,5))

plt.title("top 10 IPL venue")
plt.xlabel("venue")
plt.ylabel("match count")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

## Analysis 4: Venue Analysis

### Objective
Identify the stadiums that have hosted the most IPL matches.

### Method
Count the number of occurrences of each venue in the dataset.

### Insight
The venue with the highest match count has been the most frequently used IPL stadium.

```python
# Bonus Analysis: Top Venue by Total Runs

venue_runs = (
    ipl.groupby("venue")["runs"]
    .sum()
    .sort_values(ascending = False)
    .head(10)
)
print(venue_runs)
```

```python
# Chart
import matplotlib.pyplot as plt

venue_runs.plot(kind ="bar", figsize= (10,5))

plt.title("top 10 venue by total runs")
plt.xlabel("venue")
plt.ylabel("runs")
plt.tight_layout()

plt.show()
```

```python
# Check Available Columns:

print(ipl.columns.tolist())
```

```python
# Check the Date Format:

ipl["date"].head()
```

```python
# Convert Date to Datetime:

ipl["date"] = pd.to_datetime(ipl["date"], dayfirst=True)
```

```python
# Extract Year:

ipl["year"] = ipl["date"].dt.year

ipl[["date", "year"]].head()
```

```python
# Count Matches Per Season:

matches_per_year =(
    ipl.groupby("year")["filename"]
    .nunique()
    .sort_index()
)
print(matches_per_year)
```

```python
# Create the Chart:

import matplotlib.pyplot as plt

matches_per_year.plot (kind = "line", marker = "o", figsize = (8,5))

plt.title("IPL matcher per session")
plt.xlabel("year")
plt.ylabel("matches")

plt.show()
```

```python
#  Runs Per Season

run_per_year = (
    ipl.groupby("year")["runs"]
    .sum()
    .sort_index()
)
print(run_per_year)
```

```python
# Create the Chart

import matplotlib.pyplot as plt

run_per_year.plot(kind = "bar", figsize =(10,5))

plt.title("Total Run Scored Per Season")
plt.xlabel("year")
plt.ylabel("runs")
plt.tight_layout()

plt.show()
```
