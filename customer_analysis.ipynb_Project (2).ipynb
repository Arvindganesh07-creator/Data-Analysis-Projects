```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Arvind\Downloads\sales_analysis_project\customer_shopping_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
#  Create Total Purchase Column:
df["Total_Spent"] = df["quantity"] * df["price"]
```

```python
# Top Customers:
top_customers =(
    df.groupby("customer_id")["Total_Spent"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(top_customers)
```

```python
# Which Segments Spend Most?
segment_spending = (
    df.groupby("gender")["Total_Spent"]
    .sum()
    .sort_values (ascending=False)
)
print(segment_spending)
```

```python
# Average Customer Spending
avg_spending = (
    df.groupby("customer_id")["Total_Spent"]
    .sum()
    .mean()
)
print("Average Customer Spending:" , avg_spending)
```

```python
# Purchase Frequency
purchase_frequency= (
    df.groupby("customer_id")
    .size()
    .sort_values (ascending=False)
)
print(purchase_frequency.head())
```

```python
# Visualization:
top_customers.plot(kind= "bar")

plt.title("top 10 customer by spending")
plt.xlabel("customer id")
plt.ylabel("total spending")
plt.show()
```
